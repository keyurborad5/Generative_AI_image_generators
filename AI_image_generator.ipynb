### Part 2
    -Here We are laoding the saved model again 
    - using GPU to infere but in my system VRAM 8GB it ran out of RAM and did not work 
    - Hence I used sequencial CPU offlaod
    - And some additional memory saver which I need to learn
    - Step 4 is inferenacing by emptying the cache of cuda
    - NOTE:
        - Thing I noticed that whenever i loaded the model and I used step4 to infere my RAM was in 100% utilisation and you cannont re run this cell
        - instead you need to run only step 4 when your model is already loaded.

```python
import torch
import os
from dotenv import load_dotenv
from diffusers import FluxPipeline
'''
Step 3: once Downloaded use the model from local and no need to download again
    . Also Comment the downloading part
'''
path="./SavedModel/Flux_1-dev"
pipe = FluxPipeline.from_pretrained(path, torch_dtype=torch.bfloat16)
# OPTIONAL
try:
    pipe.enable_sequential_cpu_offload()
except Exception:
    pipe.enable_model_cpu_offload() #save some VRAM by offloading the model to CPU. Remove this if you have enough GPU power

# Additional memory savers
pipe.enable_attention_slicing("max")
pipe.enable_vae_slicing()
pipe.enable_vae_tiling()


'''
Step 4: Inference
'''
torch.cuda.empty_cache()
prompt = "A Lion holding a sign that says hello Keyur"
image = pipe(
    prompt, # the prompt to generate the image
    height=384, # height of the generated image
    width=512, # Width of the generated image
    guidance_scale=1,# how closely the image generation follows the prompt
    num_inference_steps=5, # Number of inference steps, larger the number better the image but will take more time
    max_sequence_length=256, # Max sequence length for the prompt
    generator=torch.Generator("cpu").manual_seed(0) # Seed for reproducibility
).images[0]

# Saving the images
image.save("./Images/flux-dev_5.png")
del image
torch.cuda.empty_cache()
```

### Using prelaoded Model in memory

```python
import time
for i in range(7,20):
    a=time.time()
    print("Generating image: ", i)
    prompt = "A Lion holding a sign that says hello Keyur"
    image = pipe(
        prompt, # the prompt to generate the image
        height=384, # height of the generated image
        width=512, # Width of the generated image
        guidance_scale=1,# how closely the image generation follows the prompt
        num_inference_steps=i, # Number of inference steps, larger the number better the image but will take more time
        max_sequence_length=256, # Max sequence length for the prompt
        generator=torch.Generator("cpu").manual_seed(0) # Seed for reproducibility
    ).images[0]

    # Saving the images
    image.save(f"./Images/flux-dev_{i}.png")
    del image
    b=time.time()
    print(f"Time taken for {i}inference:  {b-a}")
```

### Flux- Kontext Model

```python
import torch
import os
from dotenv import load_dotenv
from diffusers import FluxPipeline
'''
Step 3: once Downloaded use the model from local and no need to download again
    . Also Comment the downloading part
'''
path="./SavedModel/Flux_1-Kontext-dev"
pipe = FluxPipeline.from_pretrained(path, torch_dtype=torch.bfloat16)
# OPTIONAL
try:
    pipe.enable_sequential_cpu_offload()
except Exception:
    pipe.enable_model_cpu_offload() #save some VRAM by offloading the model to CPU. Remove this if you have enough GPU power

# Additional memory savers
pipe.enable_attention_slicing("max")
pipe.enable_vae_slicing()
pipe.enable_vae_tiling()


'''
Step 4: Inference
'''
for i in range(1,50,2):
    torch.cuda.empty_cache()
    prompt = "A Lion holding a sign that says \"hello Keyur\""
    image = pipe(
        prompt, # the prompt to generate the image
        height=384, # height of the generated image
        width=512, # Width of the generated image
        guidance_scale=2,# how closely the image generation follows the prompt
        num_inference_steps=i, # Number of inference steps, larger the number better the image but will take more time
        max_sequence_length=256, # Max sequence length for the prompt
        generator=torch.Generator("cpu").manual_seed(0) # Seed for reproducibility
    ).images[0]
    print(f"Generated image with inferance steps {i}: {i}")
    # Saving the images
    image.save(f"./Images/flux-kontext-dev_{i}.png")
    del image

torch.cuda.empty_cache()
```

```python
import fal_client

def on_queue_update(update):
    if isinstance(update, fal_client.InProgress):
        for log in update.logs:
           print(log["message"])

result = fal_client.subscribe(
    "fal-ai/flux/dev",
    arguments={
        "prompt": "Extreme close-up of a single tiger eye, direct frontal view. Detailed iris and pupil. Sharp focus on eye texture and color. Natural lighting to capture authentic eye shine and depth. The word \"KEYUR\" is painted over it in big, white brush strokes with visible texture."
    },
    with_logs=True,
    on_queue_update=on_queue_update,
)
print(result)
```
