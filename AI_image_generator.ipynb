```python
import torch
import os
from dotenv import load_dotenv
from diffusers import FluxPipeline

'''
Step1: huggingface_hub login
'''
from huggingface_hub import login
load_dotenv() # Getting API key from .env file , which is stored as env variable
login(token=os.getenv("HF_API_KEY"))


'''
Step 2:
Using diffusers library and importing FluxPipeline, 
Download the model and saving in local
'''
pipe = FluxPipeline.from_pretrained("black-forest-labs/FLUX.1-dev", torch_dtype=torch.bfloat16)
pipe.save_pretrained("./Saved_Model/Flux_1-dev")


'''
Step 3: once Downloaded use the model from local and no need to download again
    . Also Comment the downloading part
'''
path="./Saved_Model/Flux_1-dev"
pipe = FluxPipeline.from_pretrained(path, torch_dtype=torch.bfloat16)
# OPTIONAL
# pipe.enable_model_cpu_offload() #save some VRAM by offloading the model to CPU. Remove this if you have enough GPU power

'''
Step 4: Inference
'''
prompt = "A cat holding a sign that says hello world"
image = pipe(
    prompt, # the prompt to generate the image
    height=1024, # height of the generated image
    width=1024, # Width of the generated image
    guidance_scale=3.5,# how closely the image generation follows the prompt
    num_inference_steps=50, # Number of inference steps, larger the number better the image but will take more time
    max_sequence_length=512, # Max sequence length for the prompt
    generator=torch.Generator("cpu").manual_seed(0) # Seed for reproducibility
).images[0]

# Saving the images
image.save("./Images/flux-dev.png")
```

```python
import fal_client

def on_queue_update(update):
    if isinstance(update, fal_client.InProgress):
        for log in update.logs:
           print(log["message"])

result = fal_client.subscribe(
    "fal-ai/flux/dev",
    arguments={
        "prompt": "Extreme close-up of a single tiger eye, direct frontal view. Detailed iris and pupil. Sharp focus on eye texture and color. Natural lighting to capture authentic eye shine and depth. The word \"KEYUR\" is painted over it in big, white brush strokes with visible texture."
    },
    with_logs=True,
    on_queue_update=on_queue_update,
)
print(result)
```
